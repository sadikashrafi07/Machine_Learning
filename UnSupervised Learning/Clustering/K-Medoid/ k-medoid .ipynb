```python
import numpy as np
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt
```

```python
data = {'x': [9, 10, 4, 5, 3, 2, 8, 4, 8, 9], 'y': [6, 4, 4, 8, 8, 5, 5, 6, 4, 3]}

x = [[i, j] for i, j in zip(data['x'], data['y'])]
data_x = np.asarray(x)
```

```python
plt.scatter(data_x[:, 0], data_x[:, 1], c='blue', label='Data Points')
plt.title('Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
data_x = np.asarray(x)
data_x
```

```python
model_km = KMedoids(n_clusters=2)
```

```python
km = model_km.fit(data_x)
km
```

```python
print("Labels :",km.labels_)
print("Cluster centers  :",km.cluster_centers_)
```

```python
# Visualize the clusters
plt.scatter(data_x[model_km.labels_ == 0, 0], data_x[model_km.labels_ == 0, 1], c='red', label='Cluster 1')
plt.scatter(data_x[model_km.labels_ == 1, 0], data_x[model_km.labels_ == 1, 1], c='blue', label='Cluster 2')

# Plot cluster centers
plt.scatter(model_km.cluster_centers_[:, 0], model_km.cluster_centers_[:, 1], c='black', marker='X', s=100, label='Centroids')


plt.title('KMedoids Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
data = np.concatenate([np.random.normal(loc=10, scale=2, size=(50, 2)),
                       np.random.normal(loc=20, scale=2, size=(50, 2))])

# Apply K-Medoids
model_km = KMedoids(n_clusters=2, random_state=42)
model_km.fit(data)

# Visualize the clusters
plt.scatter(data[model_km.labels_ == 0, 0], data[model_km.labels_ == 0, 1], c='red', label='Cluster 1')
plt.scatter(data[model_km.labels_ == 1, 0], data[model_km.labels_ == 1, 1], c='blue', label='Cluster 2')

# Plot cluster centers
plt.scatter(model_km.cluster_centers_[:, 0], model_km.cluster_centers_[:, 1], c='black', marker='X', s=100, label='Medoids')

plt.title('KMedoids Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
